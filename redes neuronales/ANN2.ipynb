```python
learning_rate = 1e-3
batch_size = 256
shuffle = True


use_cuda = torch.cuda.is_available()
device = torch.device("cuda" if use_cuda else "cpu")
torch.manual_seed(0)
if use_cuda:
    torch.cuda.manual_seed(0)
```

```python
class Net(nn.Module):
    def __init__(self, input_size=16):
      super(Net, self).__init__()

      self.dropout1 = nn.Dropout(0.25)
      self.dropout2 = nn.Dropout(0.25)
      self.dropout3 = nn.Dropout(0.25)
      self.dropout4 = nn.Dropout(0.25)
      self.dropout5 = nn.Dropout(0.25)

      self.fc1 = nn.Linear(input_size, 16)
      self.fc2 = nn.Linear(16, 32)
      self.fc3 = nn.Linear(32, 64)
      self.fc4 = nn.Linear(64, 128)
      self.fc5 = nn.Linear(128, 10)
      self.fc6 = nn.Linear(10, 1)

    # x represents our data
    def forward(self, x):

      x = F.relu(self.fc1(x))
      x = self.dropout1(x)

      x = F.relu(self.fc2(x))
      x = self.dropout2(x)

      x = F.relu(self.fc3(x))
      x = self.dropout3(x)

      x = F.relu(self.fc4(x))
      x = self.dropout4(x)

      x = F.relu(self.fc5(x))
      x = self.dropout5(x)

      output = self.fc6(x)
      return output
```

```python
def load_model(PATH, optimizer, eval=False):
  model = Net(input_size=15)

  
  if use_cuda:
    checkpoint = torch.load(PATH)
  else:
    checkpoint = torch.load(PATH, map_location=torch.device('cpu'))
  model.load_state_dict(checkpoint['model_state_dict'])
  optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
  epoch = checkpoint['epoch']
  loss = checkpoint['loss']

  if eval:
    model.eval()
  else:
    model.train()
  return model.to(device), optimizer, epoch, loss
```

```python
net = Net(input_size=15).to(device)

# create a stochastic gradient descent optimizer
optimizer = optim.Adam(net.parameters(), lr=learning_rate)
```

```python
file_ = '31-1608495076.8163753.pt'#'4-1608219911.1479685.pt'
model, _, _, _ = load_model('{}/{}'.format(checkpoint_folder, file_), optimizer, eval=True)
```

```python
def eval_range(model, dataset, top=60, bottom=0):
  condition1 = dataset["Diferencia"] >= bottom
  condition2 = dataset["Diferencia"] < top
  total_cond = np.logical_and(condition1, condition2)
  filtered = dataset[total_cond]
  if filtered.shape[0] > 0:
    y = filtered["Diferencia"].values
    y_val = torch.from_numpy(y).float().to(device)#.reshape(1, -1)
    X_val = torch.from_numpy(filtered.drop(columns=["Diferencia"]).values).float().to(device)
    y_pred = model(X_val)
    y_pred = y_pred.data.cpu().numpy()
    rms = mean_squared_error(y, y_pred, squared=False)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    return rms, mae, r2, y, y_pred
  else:
    return None
```

```python
def batch_error(y, y_pred, batch=100, negative=False):
  total = 0
  error = 0
  side_error = 0
  n_batches = y.shape[0]//batch + 1
  for i in range(n_batches):
    start = i*batch
    end = (i+1)*batch
    if i == n_batches-1:
      total += y[start:].shape[0]
      dif = y[start:] - y_pred[start:]
      error += np.sum(dif)
    else:
      total += y[start:end].shape[0]
      dif = y[start:end] - y_pred[start:end]
      error += np.sum(dif)
    if negative:
      side_error += np.sum(dif[dif < 0])
    else:
      side_error += np.sum(dif[dif > 0])
  return error/total, side_error/total
```

```python
rms, mae, r2, y, y_pred = eval_range(model, val_set, top=1e20, bottom=0)
me, pos_error = batch_error(y, y_pred)
print("Errores globales")
print("\tRMS: \t{:.3f}".format(rms))
print("\tME: \t{:.3f}".format(me))
print("\tMPE: \t{:.3f}".format(pos_error))
print("\tMAE: \t{:.3f}".format(mae))
print("\tR2: \t{:.3f}".format(r2))
print('')


ranges = [[0, 5*60], [5*60, 10*60], [10*60, 20*60], [20*60, 40*60], [40*60, 60*60]]
for bottom, top in ranges:
  rms, mae, r2, y, y_pred = eval_range(model, val_set, top=top, bottom=bottom)
  me, pos_error = batch_error(y, y_pred)
  print("Errores para el rango {}min a {}min".format(bottom//60, top//60))
  print("\tRMS: \t{:.3f}".format(rms))
  print("\tME: \t{:.3f}".format(me))
  print("\tMPE: \t{:.3f}".format(pos_error))
  print("\tMAE: \t{:.3f}".format(mae))
  print("\tR2: \t{:.3f}".format(r2))
  print('')
```

```python
checkpoints = [file for file in os.listdir(checkpoint_folder) if file[-3:]=='.pt']

maes = []
rmss = []
r2s = []
idxes = []
for checkpoint in checkpoints:
  model, _, _, _ = load_model('{}/{}'.format(checkpoint_folder, checkpoint), optimizer, eval=True)
  rms, mae, r2, _, _ = eval_range(model, val_set, top=1e20, bottom=0)
  idxes.append(int(checkpoint.split("-")[0]))
  rmss.append(rms)
  maes.append(mae)
  r2s.append(r2)
```
